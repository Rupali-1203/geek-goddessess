```python
from DeepImageSearch import Load_Data, Search_Setup

# Load images from the dataset folder
dataset_path = '/content/drive/MyDrive/dataformyntra/model-dataset/Model-Dataset'  # Path to your dataset folder
image_list = Load_Data().from_folder([dataset_path])

# Set up the search engine with the 'vgg19' model
st = Search_Setup(image_list=image_list, model_name='vgg19', pretrained=True)

# Index the images
st.run_index()

# Optionally, get metadata (if needed for further processing)
metadata = st.get_image_metadata_file()

# Define the path to your query image
query_image_path = '/content/drive/MyDrive/dataformyntra/similar/GA6633.jpg'  # Replace with your query image path

# Get the top 5 similar images
similar_images = st.get_similar_images(image_path=query_image_path, number_of_images=5)

# Print the paths of the top 5 matching outfits
print("Top 5 matching outfits:")
for img_path in similar_images:
    print(img_path)

# Optionally, plot the similar images
st.plot_similar_images(image_path=query_image_path, number_of_images=5)
```
